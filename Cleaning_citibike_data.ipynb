```python
import pandas as pd
```

```python
# Read csv
Citibike_Mar2020 = pd.read_csv("Data/JC-202003-citibike-tripdata.csv")
Citibike_Mar2021 = pd.read_csv("Data/JC-202103-citibike-tripdata.csv")
```

```python
Citibike_Mar2020
```

```python
Citibike_Mar2021
```

```python
# Calculate age of riders and add age as a new column
age_2020 = []

for i in Citibike_Mar2020["birth year"]:
    age = 2020 - i
    age_2020.append(age)
```

```python
age_2021 = []

for i in Citibike_Mar2021["birth year"]:
    age = 2021 - i
    age_2021.append(age)
```

```python
Citibike_Mar2020["Age"] = age_2020
Citibike_Mar2020.head()
```

```python
Citibike_Mar2021["Age"] = age_2021
Citibike_Mar2021.head()
```

```python
# Combine both files
combined_citi = pd.concat([Citibike_Mar2020, Citibike_Mar2021])
```

```python
combined_citi
```

```python
# check data types and correct
combined_citi.dtypes
```

```python
# Change start & stoptime data type from object to datetime and gender type from integer to object
combined_citi['starttime'] = pd.to_datetime(combined_citi['starttime'])
combined_citi['stoptime'] = pd.to_datetime(combined_citi['stoptime'])
combined_citi['gender']= combined_citi['gender'].map(str)
```

```python
combined_citi.dtypes
```

```python
# Check for duplicates
duplicate_comb = combined_citi[combined_citi.duplicated()]
print(duplicate_comb)
```

```python
# Check for NAN values
combined_citi.isnull().values.any()
```

```python
# Add a new column of trip duration in mins
combined_citi['duration(mins)'] = (combined_citi['tripduration']/60).round(2)
combined_citi
```

```python
# Change gender from number to string
combined_citi.set_index(['gender'], inplace = True)
```

```python
combined_citi.rename(index = {"0": "Unknown", "1": "Male", "2": "Female"}, inplace = True)
combined_citi.head()
```

```python
combined_citi.reset_index()
```

```python
# Export into CSV
combined_citi.to_csv('Data/combined.csv', index=True)
```
